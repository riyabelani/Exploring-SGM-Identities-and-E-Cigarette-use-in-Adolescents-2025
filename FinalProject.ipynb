## Title:	E Cigarette Use among Sexual and Gender Minority Adolescents: 
California Health Interview Survey (CHIS), 2025


Summary: This study looks to explore the difference in e-cigarette usage of adolescents age 12-17 residing in California, between adolescents of sexual or gender minority (SGM) identities and adolescents of heterosexual and cisgender identities. By looking at age when e-cigarette use began, this study aims to identify potential disparities in e-cigarette usage behavior.  

### Introduction

E-cigarette use is a huge issue for adolescents and has serious health consequences. According to 2024 Annual National Youth Tobacco Survey, 1.63 million middle and high school students currently use e-cigarettes (FDA, 2024). â€¯E-cigarette contain harmful chemicals that when used by youth, can lead to serious health complications including cardiovascular disease, adolescent brain development, adolescent attention and cognition, and even death (NIH, 2016). Previous research has shown that sexual minority adolescents report higher use of e-cigarettes compared to their heterosexual peers, with bisexual, pansexual, or queer students reporting the highest prevalence of e-cigarette use (Azagba, 2025). A similar study looking at vaping among sexual and gender minority adolescents who were assigned male at birth also found increased e-cigarette use within bisexual, pansexual and queer adolescents and saw psychosocial correlations within sexual and gender minority youth (Kraus et al., 2022). A third study also found that of the 21.6% of adolescents surveyed that had a SGM identity, 10.8% reported e-cigarette use and poor mental health was a significant predictor of this use (Elexis et al., 2025). Specifically, this study and others have shown that minority stress theory explains how SGM youth may be more likely to turn to e-cigarettes for social and psychological coping compared to their cisgender and heterosexual peers, who reported using e-cigarettes for the taste or for social reasons or experimentation reasons (Mereish et al., 2022). Though these previous studies have shown an association between e-cigarette use and SGM adolescents, they all did not explore sexual and gender minorities together or have only looked at e-cigarette use as a dichotomous yes or no variable instead of looking at when the use began, which may not be the best method to analysis youth e-cigarette use, according to preliminary findings (Selya et al., 2024). Thus, this study aims to include and combine both sexual and gender identities and look at starting age of use of e-cigarettes.

Plots and outputs:
- Bar charts and boxplots were generated to compare prevalence and age at first use by SGM status and saved to a `figures` folder in Week_5.

### Discussion
Interpretation:
- Exploratory analyses indicate higher e-cigarette prevalence among adolescents classified as sexual and/or gender minorities (SGM) using both label-based and heuristic groupings. The observed earlier median age at first use in the SGM group may have implications for targeted prevention efforts.

Limitations:
-  The CHIS study uses a cross-sectional study design, which limits conclusions that are able to be made with the analysis as temporality cannot be established since the data was recorded only at one point at time. 
- The SGM classification combines multiple items and relies partly on heuristics where labels or consistent coding were not available â€” this can introduce misclassification. Likely due to the fact that the survey is for adolescents, there were certain variables that were not measured such as socioeconomic status which could have been a potential confounder. Likewise, only one dimension of sexual orientation was measured, identity. There were no questions asked about attraction and behavior, and though it may be difficult to ask adolescents about sexual behavior, it limits the credibility or verification of the sexual identity reported. 
- Some variables may use survey coding or require weighting for population-representative estimates; this notebook does not yet apply complex survey weights.

Significance:
The potential significance of this proposed analysis, should the hypothesis be supported, could lead to future research and eventually targeted programs for sexual and gender minorities combatting e-cigarette use. There are other questions on the CHIS survey that ask questions about bullying incidents at school as well as a follow up question to e-cigarette use that offers a check all that apply list of reasons as to why the adolescent uses e-cigarettes. Should the hypothesis be supported, future studies could investigate the reasoning as to what the main SGM adolescents use e-cigarettes at higher rates and how these reasons differentiate between SGM and non-SGM adolescents. Furthermore, this research could be a basis in a larger advocacy effort for targeted programs for decreasing e-cigarette use/vaping for SGM adolescents in California. Information could be provided to pediatricians for them to provide extra support or information to their sexual and gender minority patients about the health risks of e-cigarette use. Finally, continued research in this area of study could lead to increased policy around e-cigarette sales, such as more comprehensive flavor bans and stricter ID checks or less smoke shops near schools to decrease adolescent use, especially if it is shown that minority populations such as sexual and gender minorities are particularly vulnerable.

**Exploratory data analysis (EDA)**
Frequency tables for e-cigarette variables, cross-tabs by sexual orientation (`TA24`) and gender (`SRSEX_A`), and an attempted logistic regression (current vs never use).

```python
# Load mapping and dataset (reload safe for restarted kernel)
import json
import pandas as pd
mapping_path = r"C:\Users\riyabelani\Desktop\Fall 2025\GenAI\Week_5\teen_variable_mapping.json"
parquet_path = r"C:\Users\riyabelani\Desktop\Fall 2025\GenAI\Week_5\teen_cleaned.parquet"
with open(mapping_path, 'r', encoding='utf-8') as f:
    mapping = json.load(f)
print('Mapping loaded â€” variables found:')
for k,v in mapping.items():
    print(k, '->', [item['column'] for item in v])
df = pd.read_parquet(parquet_path)
print('Loaded dataframe shape =', df.shape)
```

```python
# Identify candidate e-cigarette / vaping related columns and produce frequencies and crosstabs
import numpy as np
ecig_keywords = ['ecig','e-cig','e_cig','vape','vaping','juul','e cigarette','e-cigarette']
def find_cols_by_keywords(kws, search_labels=False):
    out = []
    for c in df.columns:
        cl = c.lower()
        if any(k in cl for k in kws):
            out.append(c)
    if search_labels and 'mapping' in globals():
        for k,v in mapping.items():
            for item in v:
                lab = item.get('label','')
                if lab and any(kw in lab.lower() for kw in kws) and item['column'] not in out:
                    out.append(item['column'])
    return out
ecig_cols = find_cols_by_keywords(ecig_keywords, search_labels=True)
print('Candidate e-cigarette/vaping columns (found):', ecig_cols)
# Frequency tables
for c in ecig_cols:
    print('\n==== Variable:', c, '====')
    try:
        vc = df[c].value_counts(dropna=False).head(20)
        print(vc)
    except Exception as e:
        print('Could not compute value_counts for', c, 'error:', repr(e))
# Cross-tabs vs sexual orientation (TA24) and gender (SRSEX_A) if present
so_col = mapping.get('sexual_orientation', [{'column':'TA24'}])[0]['column'] if mapping.get('sexual_orientation') else 'TA24'
gender_col = mapping.get('self_reported_gender_of_adult', [{'column':'SRSEX_A'}])[0]['column'] if mapping.get('self_reported_gender_of_adult') else 'SRSEX_A'
print('\nUsing sexual orientation column:', so_col)
print('Using gender column:', gender_col)
for c in ecig_cols:
    print('\n--- Cross-tab for', c, 'x', so_col)
    try:
        print(pd.crosstab(df[c], df[so_col], dropna=False))
    except Exception as e:
        print('Cross-tab failed for', c, 'error:', repr(e))
    print('\n--- Cross-tab for', c, 'x', gender_col)
    try:
        print(pd.crosstab(df[c], df[gender_col], dropna=False))
    except Exception as e:
        print('Cross-tab failed for', c, 'error:', repr(e))
```

```python
# Attempt logistic regression for a detected 'current' variable (binary) vs sexual orientation + gender
import statsmodels.api as sm
from patsy import dmatrices
# Try to detect a plausible 'current' variable for e-cigarettes
current_candidates = [c for c in df.columns if 'current' in c.lower() and any(k in c.lower() for k in ['e','ecig','vape','juul','cig'])]
# Also check mapping for user-provided 'current' candidates
for item in mapping.get('current_e-cigarette_smoker', []):
    if item['column'] in df.columns and item['column'] not in current_candidates:
        current_candidates.append(item['column'])
print('Current variable candidates:', current_candidates)
if not current_candidates:
    print('No clear "current e-cigarette" variable found automatically. Skipping logistic regression. Consider selecting the correct column from the mapping printed earlier.')
else:
    # pick first candidate and coerce to binary (dropna)
    ycol = current_candidates[0]
    y = df[ycol]
    # attempt to coerce to binary: if numeric and more than 2 values, treat >0 as 1
    if pd.api.types.is_numeric_dtype(y):
        unique_vals = sorted(y.dropna().unique())
        if len(unique_vals) == 2:
            pass
        else:
            y = (y>0).astype(int)
    else:
        # convert strings like 'Yes'/'No'
        y = y.astype(str).str.lower().map(lambda s: 1 if 'yes' in s or 'current' in s or 'now' in s or 'everyday' in s or 'some' in s else 0)
    # Build predictor matrix using sexual orientation and gender (as categorical)
    X = df[[so_col, gender_col]].copy()
    # Drop rows with NA in y or predictors
    df_model = pd.concat([y, X], axis=1).dropna()
    y = df_model.iloc[:,0]
    X = df_model.iloc[:,1:]
    # Create dummies
    X = pd.get_dummies(X, drop_first=True)
    X = sm.add_constant(X)
    print('Fitting logistic regression with outcome', ycol, 'on', X.shape[1]-1, 'predictors and', len(y), 'observations')
    try:
        model = sm.Logit(y, X).fit(disp=False)
        print(model.summary())
    except Exception as e:
        print('Logistic regression failed:', repr(e))
```

```python
# Create SGM (sexual & gender minority) grouping using TA24, SRSEX_A, TL37V2
import pyreadstat
sas_path = r"C:\Users\riyabelani\Desktop\Fall 2025\GenAI\Week_5\teen.sas7bdat"
# Read SAS metadata (fast)
_, sas_meta = pyreadstat.read_sas7bdat(sas_path, usecols=[])
# Helper: get value labels dict for a variable name (map code->label)
def get_val_labels(var):
    v = {}
    try:
        v = sas_meta.value_labels.get(var, {})
    except Exception:
        v = sas_meta.variable_value_labels.get(var, {}) if hasattr(sas_meta, 'variable_value_labels') else {}
    return v
# Variables of interest (names discovered earlier)
so_var = 'TA24'
gender_var = 'SRSEX_A'
gender_expr_var = 'TL37V2'  # may exist
# Extract labels (if available)
so_labels = get_val_labels(so_var)
gender_labels = get_val_labels(gender_var)
gexpr_labels = get_val_labels(gender_expr_var)
print('Sexual orientation labels (sample):', list(so_labels.items())[:10])
print('Gender labels (sample):', list(gender_labels.items())[:10])
print('Gender expression labels (sample):', list(gexpr_labels.items())[:10])
# Load the cleaned dataframe if not already loaded
import pandas as pd
parquet_path = r"C:\Users\riyabelani\Desktop\Fall 2025\GenAI\Week_5\teen_cleaned.parquet"
df = pd.read_parquet(parquet_path)
# Function to determine if a label indicates minority
def label_is_minority(lbl):
    if not lbl or not isinstance(lbl, str):
        return False
    s = lbl.lower()
    if 'hetero' in s or 'straight' in s or 'heterosexual' in s:
        return False
    if any(x in s for x in ['missing','refus','dont know','refused','not asked','unknown']):
        return False
    if any(x in s for x in ['gay','lesbian','bisex','bi-','queer','pan','asexual','trans','non-binary','nonbinary','genderqueer','other']):
        return True
    return False
# Build mappings from code->is_minority for each variable using their labels
so_minority_codes = {code: label_is_minority(lbl) for code,lbl in so_labels.items()}
gender_minority_codes = {code: label_is_minority(lbl) for code,lbl in gender_labels.items()}
gexpr_minority_codes = {code: label_is_minority(lbl) for code,lbl in gexpr_labels.items()}
print('SO minority codes (sample):', list(so_minority_codes.items())[:10])
# Apply to dataframe: create human-readable label columns (if labels exist) and sgm_flag
def map_label_series(series, val_labels):
    out = series.copy()
    m = {}
    for k,v in val_labels.items():
        m[str(k)] = v
    def mapper(x):
        if pd.isna(x):
            return None
        return m.get(str(x), str(x))
    return out.map(mapper)
# Add label columns (if possible)
if so_labels:
    df['so_label'] = map_label_series(df[so_var], so_labels)
else:
    df['so_label'] = df[so_var].astype(str)
if gender_labels:
    df['gender_label'] = map_label_series(df[gender_var], gender_labels)
else:
    df['gender_label'] = df[gender_var].astype(str)
if gexpr_labels:
    df['gexpr_label'] = map_label_series(df[gender_expr_var], gexpr_labels)
else:
    df['gexpr_label'] = df.get(gender_expr_var, pd.Series([None]*len(df))).astype(str)
# Create boolean columns indicating minority by each variable
def is_min_by_code(col, code_map):
    def f(x):
        if pd.isna(x):
            return False
        return bool(code_map.get(x, code_map.get(str(x), False)))
    return df[col].apply(f)
df['_so_min'] = is_min_by_code(so_var, so_minority_codes)
df['_gender_min'] = is_min_by_code(gender_var, gender_minority_codes)
# For gender expression, if any label flagged minority, use that
df['_gexpr_min'] = False
if gexpr_labels:
    df['_gexpr_min'] = df[gender_expr_var].apply(lambda x: bool(gexpr_minority_codes.get(x, gexpr_minority_codes.get(str(x), False))))
# Define overall SGM flag: any of the three is True
df['sgm_flag'] = df[['_so_min','_gender_min','_gexpr_min']].any(axis=1)
print('SGM flag created: counts ->')
print(df['sgm_flag'].value_counts(dropna=False))
```

```python
# Heuristic SGM grouping (majority-code approach) and comparisons
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(style='whitegrid')

# Variables
so = 'TA24'
gender = 'SRSEX_A'
gexpr = 'TL37V2'
ever_col = 'TE79'
current_col = 'ESMKCUR'
age_first = 'TE82_P1'
distress = 'DSTRS12'
# Find 'reason targeted for hate incident witnessed: sexual orientation' by label search
label_map = sas_meta.column_names_to_labels if hasattr(sas_meta, 'column_names_to_labels') else {}
reason_col = None
for col, lab in label_map.items():
    if isinstance(lab, str) and 'REASON PERSON TARGETED' in lab.upper() and 'SEXUAL ORIENTATION' in lab.upper():
        reason_col = col
        break
# fallback: look for 'REASON' and 'HATE' and 'SEXUAL' in label
if reason_col is None:
    for col, lab in label_map.items():
        if isinstance(lab, str) and 'REASON' in lab.upper() and 'HATE' in lab.upper() and 'SEXUAL' in lab.upper():
            reason_col = col
            break
print('Detected reason_col:', reason_col)

# Create sexual minority flag: treat most common non-null TA24 value as 'heterosexual'
ta_counts = df[so].value_counts(dropna=True)
if len(ta_counts)==0:
    raise RuntimeError('No values for sexual orientation variable')
hetero_code = ta_counts.index[0]
print('Assuming TA24 hetero code =', hetero_code)
so_min = df[so].notna() & (df[so] != hetero_code)
# Gender minority: if gender has >2 categories, treat non-mode as minority
g_counts = df[gender].value_counts(dropna=True)
gender_min = pd.Series(False, index=df.index)
if len(g_counts) > 2:
    gmode = g_counts.index[0]
    gender_min = df[gender].notna() & (df[gender] != gmode)
else:
    # check for explicit transgender/nonbinary codes in gexpr if available
    gender_min = pd.Series(False, index=df.index)
# Gender expression minority: non-mode -> minority
if gexpr in df.columns:
    ge_counts = df[gexpr].value_counts(dropna=True)
    if len(ge_counts)>0:
        gemode = ge_counts.index[0]
        gexpr_min = df[gexpr].notna() & (df[gexpr] != gemode)
    else:
        gexpr_min = pd.Series(False, index=df.index)
else:
    gexpr_min = pd.Series(False, index=df.index)

# Combined SGM flag
df['sgm_flag_heuristic'] = so_min | gender_min | gexpr_min
print('SGM counts (heuristic):')
print(df['sgm_flag_heuristic'].value_counts(dropna=False))

# Helper to compute proportions and chi2
from statsmodels.stats.proportion import proportions_ztest

def prop_table(col):
    tab = pd.crosstab(df['sgm_flag_heuristic'], df[col], dropna=False)
    return tab

# Compute and show tables for ever/current/distress/reason
cols_to_compare = [ever_col, current_col, age_first, distress]
if reason_col:
    cols_to_compare.append(reason_col)

summary_results = {}
for c in cols_to_compare:
    if c not in df.columns:
        print('Column not found:', c)
        continue
    print('\n=== Variable:', c, '===')
    print(df[c].value_counts(dropna=False).head(20))
    tab = pd.crosstab(df['sgm_flag_heuristic'], df[c], dropna=False)
    print('\nCross-tab:')
    print(tab)
    # compute percent positives if binary-like: pick the smaller of two unique non-null values
    vals = df[c].dropna().unique()
    if len(vals)>0 and len(vals)<=10:
        # try to compute proportion of first non-null value (assuming coding 1=yes)
        try:
            # find a 'positive' code heuristically: pick the value with smaller count (assumed 'yes')
            counts = df[c].value_counts(dropna=True)
            pos_val = counts.index[0] if counts.index[0] != -1 else (counts.index[1] if len(counts)>1 else counts.index[0])
            # compute proportions where column == pos_val
            grp = df.groupby('sgm_flag_heuristic')[c].apply(lambda s: (s==pos_val).sum())
            n = df['sgm_flag_heuristic'].value_counts()
            print('\nCounts of pos_val (assumed):', pos_val)
            print(grp)
            # z-test
            try:
                count = grp.values
                nobs = [n.loc[False], n.loc[True]]
                stat, pval = proportions_ztest(count, nobs)
                print('proportions_ztest stat=%.3f p=%.4f' % (stat, pval))
            except Exception as e:
                pass
        except Exception as e:
            pass

# Plotting: prevalence bar charts for ever and current
fig_dir = r"C:\Users\riyabelani\Desktop\Fall 2025\GenAI\Week_5\figures"
import os
os.makedirs(fig_dir, exist_ok=True)

for c,label in [(ever_col,'Ever used e-cigarettes'), (current_col,'Current e-cigarette use'), (distress,'Psychological distress last year')]:
    if c not in df.columns:
        continue
    plot_df = df[[c,'sgm_flag_heuristic']].copy()
    # make binary indicator for the smaller value
    vals = plot_df[c].dropna().unique()
    if len(vals)==0:
        continue
    counts = plot_df[c].value_counts(dropna=True)
    pos = counts.index[0]
    plot_df['pos'] = (plot_df[c]==pos).astype(int)
    agg = plot_df.groupby('sgm_flag_heuristic')['pos'].mean().reset_index()
    agg['sgm'] = agg['sgm_flag_heuristic'].map({False:'Non-SGM', True:'SGM'})
    plt.figure(figsize=(5,4))
    sns.barplot(x='sgm', y='pos', data=agg)
    plt.ylim(0,1)
    plt.ylabel('Proportion')
    plt.title(label+ ' by SGM status')
    out_file = os.path.join(fig_dir, f'{c}_by_sgm.png')
    plt.savefig(out_file, dpi=150, bbox_inches='tight')
    plt.show()
    print('Saved', out_file)

# Age when first tried: show among those who ever tried (exclude -1 and missing)
if age_first in df.columns:
    age = df.loc[df[ever_col]==1, age_first] if ever_col in df.columns else df[age_first]
    # But coding might be different; instead select age values > 0 and not equal to -1
    age = df[age_first]
    valid = age.dropna()
    valid = valid[valid>0]
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df['sgm_flag_heuristic'].map({False:'Non-SGM', True:'SGM'}), y=valid)
    plt.title('Age when first tried e-cigarettes by SGM status')
    out_file = os.path.join(fig_dir, 'age_first_by_sgm.png')
    plt.savefig(out_file, dpi=150, bbox_inches='tight')
    plt.show()
    print('Saved', out_file)

print('\nFinished comparisons and saved figures to', fig_dir)
```

```python
# Compute accurate numeric comparisons using explicit code=1 as 'Yes' (where applicable)
variables = {
    'ever_used': 'TE79',
    'current_use': 'ESMKCUR',
    'age_first': 'TE82_P1',
    'distress': 'DSTRS12'
}
# reason for hate: try TD74_2 if present, else TD66 or TD71
reason_candidates = ['TD74_2','TD66','TD71']
reason_col = None
for r in reason_candidates:
    if r in df.columns:
        reason_col = r
        break

def proportion_by_sgm(col, positive_value=1):
    tab = pd.crosstab(df['sgm_flag_heuristic'], df[col], dropna=False)
    # compute proportion of rows with value == positive_value for each group
    res = {}
    for grp in [False, True]:
        grp_idx = df['sgm_flag_heuristic'] == grp
        total = grp_idx.sum()
        if total == 0:
            res[grp] = (0,0,0)
            continue
        pos = ((df.loc[grp_idx, col] == positive_value).sum())
        res[grp] = (pos, total, pos/total)
    return res

results = {}
for name, col in variables.items():
    if col not in df.columns:
        print('Missing column', col)
        continue
    results[name] = proportion_by_sgm(col, positive_value=1)

if reason_col:
    results['reason_targeted_sexual_orientation'] = proportion_by_sgm(reason_col, positive_value=1)

# Print results
for k,v in results.items():
    print('---', k)
    for grp in [False, True]:
        pos,total,prop = v[grp]
        label = 'Non-SGM' if grp==False else 'SGM'
        print(f"{label}: {pos}/{total} = {prop:.3%}")

# Median age when first tried among those who have ever used (TE79==1)
if variables['age_first'] in df.columns and variables['ever_used'] in df.columns:
    agecol = variables['age_first']
    mask_ever = df[variables['ever_used']] == 1
    medians = {}
    for grp in [False, True]:
        grp_mask = df['sgm_flag_heuristic'] == grp
        sel = df.loc[mask_ever & grp_mask, agecol]
        sel = sel.dropna()
        sel = sel[sel>0]
        if len(sel)>0:
            medians[grp] = sel.median()
        else:
            medians[grp] = None
    print('\nMedian age when first tried (among ever users):')
    for grp in [False, True]:
        label = 'Non-SGM' if grp==False else 'SGM'
        print(f"{label}: {medians[grp]}")
```
